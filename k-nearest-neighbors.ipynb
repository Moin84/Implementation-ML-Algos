```python
import pandas as pd

data = pd.read_csv("/kaggle/input/knn-dataset/knn_dataset.txt", header=None)
data.head()
```

```python
col_names = ['Id', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'I', 'J', 'Class']

data.columns = col_names
data.columns
```

```python
data = data.drop(['Id', 'F'], axis=1)
```

```python
data.isnull().sum()
data.dtypes
```

```python
train = data.drop(['Class'], axis=1)
test = data['Class']

from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(train, test, test_size=0.2, random_state=42)
train_x.head()
```

```python
cols = train_x.columns

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)

train_x = pd.DataFrame(train_x, columns=[cols])
test_x = pd.DataFrame(test_x, columns=[cols])

train_x
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(train_x, train_y)
pred_y = model.predict(test_x)

pred_y
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score

errors = mean_absolute_error(test_y, pred_y)
errors
```

```python
for k in range(1, 6):
    model = KNeighborsClassifier(k)
    model.fit(train_x, train_y)
    pred_y = model.predict(test_x)

    accuracy = accuracy_score(pred_y, test_y)
    print("Accuracy for", k, " : ", accuracy)

    
```
