```python
import pandas as pd

data = pd.read_csv("/kaggle/input/cars-evaluation-dt/car_evaluation.csv")
data.head()
```

```python
data.columns = ["buying price", "maintenance cost", "number of doors", "number of persons", "lug_boot", "safety", "decision"]
data.head()
```

```python
data.isnull().sum()
```

```python
train = data.drop(['decision'], axis=1)
test = data['decision']
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(train, test, test_size=0.2, random_state=42)
train_x.head()
```

```python
from category_encoders import OrdinalEncoder

encoder = OrdinalEncoder(cols=["buying price", "maintenance cost", "number of doors", "number of persons", "lug_boot", "safety"])
train_x = encoder.fit_transform(train_x)
test_x = encoder.transform(test_x)

train_x
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=6, random_state=42)
model.fit(train_x, train_y)
pred_y = model.predict(test_x)
pred_y
```

```python
from sklearn.metrics import accuracy_score

for d in range(1, 11):
    model = DecisionTreeClassifier(max_depth=d, random_state=42)
    model.fit(train_x, train_y)
    pred_y = model.predict(test_x)

    print("Accuracy for", d, "depth", accuracy_score(test_y, pred_y))
```
