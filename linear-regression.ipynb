```python
import pandas as pd

data = pd.read_csv("/kaggle/input/startup-logistic-regression/50_Startups.csv")
data.head()
```

```python
train = data.drop(["Profit", "State"], axis=1)
test = data["Profit"]

from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(train, test, test_size=0.2, random_state=42)
train_x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_x, train_y)
pred_y = model.predict(test_x)
pred_y
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test_y, pred_y)
print("Mean Absolute Error: ", error)
```
