```python
import pandas as pd

data = pd.read_csv("/kaggle/input/diabetes-dataset/diabetes2.csv")
data.head()
```

```python
train = data.drop(["Outcome"], axis=1)
test = data['Outcome']
train
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(train, test, test_size=0.2)
train_x.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train_x, train_y)
pred_y = model.predict(test_x)
pred_y
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test_y, pred_y)
print("Mean Absolute Error: ", error)
```
